```python
#Defining a parent class
class Marsupial:
    def __init__(self):
        self.array = []
     #Define a method to append values to the initialized array abobe   
    def put_in_pouch(self, itemName):
        self.array.append(itemName)
    #Method to print the values from an array
    def pouch_contents(self):
        print(self.array)
        
 #Defining a child class       
class Kangaroo(Marsupial):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        super().__init__()
   
    def jump(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"
        
        
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()

k = Kangaroo(0,0)
print(k)

k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()  

k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)

```

```python
def binary(n, intbinaryrep = ""):
    if n > 1:
        res = n % 2
        intbinaryrep = str(res) + intbinaryrep
        binary(int(n/2), intbinaryrep)
    else:
        print(f"binary({n})")
        print(str(n) + intbinaryrep)
        
binary(0)
binary(1)
binary(3)
binary(9)
```

```python
from html.parser import HTMLParser

#class created with HTML Parser as input param
class HeadingParser(HTMLParser):
    # Initializing variable indentation and setting te value to -1
    indentation = -1
    
    # Define a class that expects the Html tag to validate for the condition
    def handle_starttag(self, tag, attrs):
        # validating tags to assign specific indentaion values
        if tag == 'h1':
            self.indentation = 0
        if tag == 'h2':
            self.indentation = 1
        # multiple elif's can be added to check for other tags as well to assign indentation values to the header
        else:
            self.indentation = -1
    
    # method removing trailing spaces
    def handle_data(self, data):
        if data.strip() != "":
            # invoking to print the indented text
            Indentatedtext(data, self.indentation)
        

def Indentatedtext(text, indentation):
    if indentation >= 0:
        print(f"{' ' * indentation}{text}")        

#Opening the html file
infile = open('w3c.html') 
content = infile.read() 
infile.close() 

headingParser = HeadingParser()
headingParser.feed(content)
```
